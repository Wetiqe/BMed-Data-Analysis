# Analysis
患者前后测行为评分对比

```python
pt12_contrast = []
result_table = []
scales = pre_test.columns[1:]

for scale in scales:
    perm_paired_t = robjects.r['perm_matched_t'](r.to_csv_r(pre_test, 'pre', scale), r.to_csv_r(post_test, 'post', scale), 4 + len(scale), nreps=nreps)
    
    pt12_contrast.append(stats.shapiro(pre_test[scale])+stats.shapiro(post_test[scale])+stats.ttest_rel(pre_test[scale],post_test[scale])+tuple(perm_paired_t))

pt12_contrast_df = pd.DataFrame(pt12_contrast, index=scales, columns=['pre_stat','pre_p', 'post_stat','post_p','paired_t','paired_p','perm_paired_p'])
```
